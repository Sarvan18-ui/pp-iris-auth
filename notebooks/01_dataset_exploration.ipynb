```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

dataset_path = "../data/CASIA/CASIA-Iris-Thousand"
csv_path = "../data/CASIA/iris_thousands.csv"

df = pd.read_csv(csv_path)

df.head()
```

```python
print("Total images:", len(df))
```

```python
print("Total subjects:", df["Label"].nunique())
```

```python
import os

possible_paths = [
    "../data/CASIA/CASIA-Iris-Thousand",
    "../data/CASIA/CASIA-Iris-Thousand/thousand_iris",
    "../data/CASIA/thousand_iris"
]

for path in possible_paths:
    print(path, "->", os.path.exists(path))
```

```python

import glob

fig, axes = plt.subplots(2, 3, figsize=(12, 8))

for i, ax in enumerate(axes.flat):
    original_path = df.iloc[i]["ImagePath"]
    filename = os.path.basename(original_path)

    search_pattern = os.path.join(dataset_path, "**", filename)
    matches = glob.glob(search_pattern, recursive=True)

    if len(matches) == 0:
        ax.set_title("Image not found")
        ax.axis("off")
        continue

    img_path = matches[0]

    img = Image.open(img_path)
    ax.imshow(img, cmap="gray")

    ax.set_title(f"Label: {df.iloc[i]['Label']}")
    ax.axis("off")

plt.tight_layout()
plt.show()
```

```python
subject_counts = df["Label"].value_counts()

plt.figure(figsize=(12,5))

subject_counts.head(20).plot(kind='bar')

plt.title("Image Distribution for First 20 Subjects")
plt.xlabel("Subject Label")
plt.ylabel("Number of Images")

plt.xticks(rotation=45)

plt.show()
```

```python
subject_counts = df["Label"].value_counts()

plt.figure(figsize=(12,5))

plt.hist(subject_counts, bins=20)

plt.title("Subject Distribution Graph")
plt.xlabel("Number of Images per Subject")
plt.ylabel("Number of Subjects")

plt.show()
```
